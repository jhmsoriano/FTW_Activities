```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
world = pd.read_excel('/content/Data_Extract_From_World_Development_Indicators.xlsx', sheet_name='Data')
print(world.head())
```

```python
world.info()
```

```python
world[world.isna().any(axis=1)]
```

```python
world.drop(world.index[820:], inplace=True)
world.info()
```

```python
world_ex = world.loc[world['Series Name'].isin(['Current health expenditure (% of GDP)','Population growth (annual %)'])]
world_ex
```

```python
world_ex_drop = world_ex.drop(world_ex.columns[[1,3,13,14]], axis = 1)
world_ex_drop
```

```python
world_years = world_ex_drop.rename(columns={'2010 [YR2010]': 2010, '2011 [YR2011]': 2011, '2012 [YR2012]': 2012, '2013 [YR2013]': 2013, '2014 [YR2014]': 2014, '2015 [YR2015]': 2015, '2016 [YR2016]': 2016, '2017 [YR2017]': 2017, '2018 [YR2018]': 2018,})
world_years
```

```python
Philippines = world_years.loc[world_years["Country Name"] == "Philippines"]
Philippines
```

```python
Philippines_transposed = Philippines.T
Philippines_transposed
```

```python
Philippines_subset = Philippines_transposed[2:]
Philippines_subset
```

```python
# create a figure and axis
fig, ax = plt.subplots()

# scatter the Current health expenditure (% of GDP) against the Population growth (annual %)
ax.scatter(Philippines_subset[451], Philippines_subset[481])

# set a title and labels
ax.set_title('Philippines Dataset')
ax.set_xlabel('Current health expenditure (% of GDP)')
ax.set_ylabel('Population growth (annual %)')

plt.show()
```

1. There is a weak correlation between Health Expenditures and Population Growth from the Philippines dataset from 2010 to 2018.

```python
columns = Philippines_subset.columns

fig, ax = plt.subplots()
for column in columns:
    ax.plot(Philippines_subset[column])
ax.set_title('Philippines Dataset')
ax.legend(Philippines_subset[columns])
```

2. From 2010 to 2018, Philippine health expenditures (451) have remained within 4.0% to 4.5% threshold.
3. Meanwhile, Population growth in the Philippines (481) has steadily dropped from 2010 to 2018.
